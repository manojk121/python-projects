# Total Sales

```python
df = pd.merge(left = pizza, right = od, on = "pizza_id" , how = "right")
df
```

```python
df['amount'] = df['price']*df['quantity']
df
```

```python
print('total sales generated is', df['amount'].sum())
```

# most ordered pizza

```python
df1 = pd.merge(left = df, right = pt, on = "pizza_type_id", how = "left")
df1
```

```python
gb = df1.groupby('name').agg({"order_details_id":"count"})
gb
```

```python
mostorder = gb["order_details_id"].idxmax()
print("most ordered pizza is", mostorder)
```

```python
gb1 = df1.groupby("name").agg({"amount":"sum"})
gb1
```

```python
bs = gb1["amount"].idxmax()     #bs = best selling
bs
gb1.loc[bs]
```

# which pizza category is hotselling

```python
gb1 = df1.groupby("category").agg({"amount":"sum"})
gb1
```

```python
plt.pie(x=gb1["amount"], labels = gb1.index, autopct = "%1.2f%%")
plt.show()
```

# max high order time & min low order time

```python
orders
```

```python
orders['hour'] = orders['time'].dt.strftime("%H")
orders
```

```python
gb2 = orders.groupby("hour").agg({"order_id":"count"})
gb2
```

```python
sns.lineplot(x = gb2.index, y = gb2["order_id"])
plt.show()
```

```python
max = gb2['order_id'].idxmax()
min = gb2['order_id'].idxmin()

print("bussiest hour: ",max)
print("idle hour: ", min)
```

# which time people getting more pizzas

```python
orders['hour'] = orders['hour'].astype("int")
```

```python
orders.loc[(orders["hour"] < 12), "shift"] = "morning"
orders.loc[((orders["hour"] >= 12)&(orders['hour'] < 16)), "shift"] = "afternoon"
orders.loc[((orders["hour"] >= 16)&(orders['hour'] < 21)), "shift"] = "evening"
orders.loc[orders["hour"]>= 21, "shift"] = "night"
```

```python
orders
```

```python
sns.countplot(x = 'shift', data = orders)
```
